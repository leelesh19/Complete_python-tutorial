```python
#tuple are immutable but we use some methods like .count and .index
# packing and unpacking of tuple

packed_tuple = 2, 3.14, 'oyee', True
print(packed_tuple)
```

```python
# Unpacking
a, b, c, d = packed_tuple
print(a); print(b); print(c); print(d)
```

```python
# Nested tuple

lst = [(2,3,5,2), ['a', 'b', 'c'], {3, 3.14, True, False}]

print(lst)
print(lst[1][2])
print(lst[2])
```

```python
# Iterating over nested 
for sub_lst in lst:
    for item in sub_lst:
        print(item, end = ' ')
    print( )
```

## SETS

```python
# Using some oparetion on sets
my_set= {1,2,3,4,6,8,4,23,2,2,1,1,2,1,4,2,1,2}

my_set.pop()
print(my_set)
print(1 in my_set)
print(3 in my_set)
print(56 in my_set)
print(2 in my_set)
```

```python
set1 = {2, 44, 22, 11, 33, 2, 66, 00, 99}
set2 = {55, 1,3, 44, 55,77, 88, 1011, 4, 11, 2, 0}

I = set1.intersection(set2)
U = set1.union(set2)

print(f"set1 | set2 = {sorted(U)}")
print(f"set1 & set2 = {I}")

set1.intersection_update(set2)  # changed set1 element to common ele of set2 
print(set1); print(set2)
```

```python
# difference of sets
set1 = {2, 44, 22, 11, 33, 2, 66, 00, 99}
set2 = {55, 1,3, 44, 55,77, 88, 1011, 4, 11, 2, 0}

print(set1.difference(set2)) # its set1 - set2 (only set1)
```

```python
# A|B - A&B

set1.symmetric_difference(set2)
```

```python
# Sets methods

# is subset
s1 = {1,2,4,3,4,5,7,6,8,9}
s2= {1,2,3}
print(s1.issubset(set2))

# is superset
print(s1.issuperset(s2))
```

```python
lst = [2, 3,4,2,3, 2,3,2,1,2]
print(list(set(lst)))    # remove all duplicate element in list
```

```python
# counting unique words in txt

txt = "In This Python Code We Are Learing About Sets"
words = txt.split()
unique = set(words)
print(sorted(unique))
print(len(unique))
```
