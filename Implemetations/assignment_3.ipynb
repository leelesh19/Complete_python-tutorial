```python
# make random list and do some code with function

import random

random_num = [random.randint(1, 30) for _ in range(15)]

print("The random element list : ", random_num)

assend_num = sorted(random_num)
print(f"The assending list of random : {assend_num}")

dessending_num = sorted(random_num, reverse=True)
print(f"The dessending list of random : {dessending_num}")

unique_num = list(set(random_num))
print(f"list with duplicates removed : {unique_num}")
```

```python
# make dict of student on their score in decending

student = [
    {"name" : "anshu", "score" : 68},
    {"name" : "ishi", "score" : 79},
    {"name" : "lakhan-2", "score" : 86},
    {"name" : "lakhan", "score" : 91},
    {"name" : "nihu", "score" : 75}
]

sorted_std = sorted(student, key=lambda x : x["score"], reverse=True)

print("Sorted student in decending order of their score :")
for i in sorted_std:
    print(i)
```

```python
# make a matrix and their transposed
def transpose_matrix(matrix):
    transposed = [[matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix[0]))]
    return transposed

matrix = [
    [2, 4, 6],
    [3, 6, 7],
    [30, 50, 60,]
]

transposed = transpose_matrix(matrix)

print("Original matrix :")
for row in matrix:
    print(row)

print("Transposed matrix :")
for row in transposed:
    print(row)
```

```python
# normal and flattens list 
def flattens_list(nested_list):
    flat_list = [item for sublist in nested_list for item in sublist]
    return flat_list



nested_list= [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
flattened = flattens_list(nested_list)

print("Original list :")
print(nested_list)

print("Flattened list :")
print(flattened)
```

```python
# combine two list of tuple as zipped 

list1 = [1, 2, 5, 6, 9]
list2 = ['a', 'b', 'c', 'd', 'e']

zipped = list(zip(list1, list2))
print(zipped)
```

```python
# make a new list of old list in reverse order 
def reverse_list(lst):
    return lst[::-1]

lst = ['a', 'b', 'c', 'd']

reversed_lst = reverse_list(lst)
print("Original list :", lst)
print(f"Reversed list {reversed_lst}")
```

```python
# rotate a list in n times 
def rotate_lst(lst, n):
    return lst[n:] + lst[:n]

lst = [1, 2, 4, 7, 9, 33]
rotated = rotate_lst(lst, 3)

print(f"Original list : {lst}")
print(f"Rotated list : {rotated}")
```

```python
# get intersection of two list
def inter_lst(i, j):
    return [x for x in i if x in j]

lst1 = [2, 45, 66, 22, 3, 5, 7, 23]
lst2 = [3, 5, 24, 5, 2, 54, 74, 38]

intersection = inter_lst(lst1, lst2)

print(f"list1 : {lst1}")
print(f"list2 : {lst2}")
print(f"Intersection : {intersection}")
```
