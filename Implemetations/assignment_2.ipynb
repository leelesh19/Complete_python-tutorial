```python
## fibonacci sequence 
n = int(input("Enter a number for fabinacci sequence : "))
a, b = 0, 1
count = 0 
while count<n: 
    a, b = b, a+b
    count += 1
    print(a)
print(f"The given input for fabinacci seq. is {n} times")
```

```python
## factorial calculate

num = int(input("Enter a num : "))

factorial = 1
i = 1

while i<=num:
    factorial *= i ## factorial = factorial * i
    i += 1
print(f"The factorial of {num} is {factorial}.")
```

```python
## find sum of digit of input
num = int(input("Enter a num : "))

sum_of_digit = 0
while num>0:
    digit = num %10
    sum_of_digit+=digit
    num = num//10 
print(f"The sum of {num} digits is {sum_of_digit}")
```

```python
##prime number check 
num=int(input("Enter a num :  "))
is_prime = True 
if num<=1:
    is_prime=False
else : 
    for i in range(2, int(num**0.5)+1):
        if num%i == 0:
            is_prime=False
            break

if is_prime:
    print(f"{num} is a prime number.")
else :
    print(f"{num} is not a prime number.")
```
