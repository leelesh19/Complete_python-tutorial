```python
# set oparetion
set1 = set(range(1, 6))
set2 = set(range(0, 11, 2))

print(f"A = {set1}, B = {set2}")
print(f"Union = {set1.union(set2)}")
print(f"Intersection = {set1.intersection(set2)}")
print(f"Difference = {set1.difference(set2)}")
print(f"U of !I = {set1.symmetric_difference(set2)}")
```

```python
# subset and superset
import random 

set1 = set(random.randint(1, 7) for _ in range(28))
set2 = set(random.randint(1, 15) for _ in range(40))

print(f"A = {set1}, B = {set2}")
print(f"Is Subset = {set1.issubset(set2)}")
print(f"Is Superset = {set1.issuperset(set2)}")
print(f"Is Subset = {set2.issubset(set1)}")
print(f"Is Superset = {set2.issuperset(set1)}")
```

# DICT ASSIGNMENT

```python
# create a dict with first int key and their square
sq = {i:i**2 for i in range(1,20,2)}
print(sq)
```

```python
print(f"Accessing value of key 5 element : {sq[5]}")
print(f"All keys of dict : {list(sq.keys())}")
sq[20]=400
del sq[1]
print(f"Modified dict : {sq}")
```

```python
# Iterate the dict element
for i,j in sq.items():
    print(f"Key - {i}, Value - {j}")
```

```python
print({x:x**3 for x in sorted((1, 0, 9, 11, 2.5, 5, 100, 0.0, 7))})
```

```python
# merging two dict
sq2 = {p:p**2 for p in range(11, 30, 2)}
sq.update(sq2)
for i,j in sq.items():
    print(f"Key - {i}, Value - {j}")
```

```python
# Nested dict
n_dict = {
    "fr":{
        "me",
        "bro"
    },
    "expectation":{"family+me"},
    "skill_ability":{"wanna to do"},
    "learn_lesson":{
        "I",
        "bro",
        "lord_krishna"
    }

}

print(n_dict)
```

```python
#Iterate nested dict
for n_k,n_v in n_dict.items():
    print(f"{n_k} : {n_v}")
    for k in n_v:
        print(f"Q - {n_k}, ANS - {k}")

```

```python
# dict of list
dict_tpl = {i:[i*j for j in range(0, 6)] for i in range(0, 9, 2)}
for i, j in dict_tpl.items():
    print(f"THE MULTIPLE OF {i} is : {j}")
print(f"Dict_list : {dict_tpl}")
```

```python
# dict of tuple
dict_tpl = {i:(i, i*i) for i in range(0, 9)}
for i, j in dict_tpl.items():
    print(f"THE SQUARE OF {i} is : {j}")
print(f"Dict_tp : {dict_tpl}")
```

```python
# defualt dict
from collections import defaultdict

default_d = defaultdict(list)
default_d["a"].append(1)
default_d["a"].append(2)
default_d["b"].append(3)
default_d["z"].append(0)

print(default_d)
```

```python
# counting with dictonary
def count_chars(s):
    count_dict = {}
    for char in s:
        count_dict[char]=count_dict.get(char, 0) + 1
    return count_dict

string = "hii python "
print(count_chars(string))
```
