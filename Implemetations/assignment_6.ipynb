# Function ASSIGNMENT :

```python
# simple function
x = [2,5,7,9,11]
def cube(x):
    return x**3
print(list(map(cube, x)))
```

```python
# Multi iterable
def multi(x, y):
    return x*y
print(f"x*y = {multi(91, 5)}")
```

```python
# Defualt function
def multi(x, y=600):
    return (x*100)/y
print(f"10th % is = {multi(520)}")
```

```python
# keyword args - dict

def format_str(first_name, wants, interest):
    return f"{first_name} have interest in {interest} but now just wants {wants}..."

print(format_str(first_name="remo", wants="to be worthy", interest="coding"))
print(format_str(first_name="B_temporary", wants="the peace", interest="to find people"))
print(format_str(first_name="I", wants="to face everythigs", interest="to explore myself"))
```

```python
# star_arguments - apply code in unnumberable input
def positional_args(*args):
    multi_each = 1
    for n in args:
        multi_each *= n
    return multi_each
    
print(f"* each n with 1 is : {positional_args(2, 3, 5, 7.5)}")
print(f"* each n with 1 is : {positional_args(5, 5, 25, 1, 1.5, 0.5)}")
```

```python
# Nested function

def outer(x, y):
    def inner(a, b):
        return a**b
    print(f"{x}*{y} = {x*y}"); print(f"{x}+{y} = {x+y}"); print(f"{x}-{y} = {x-y}"); print(f"{x}//{y} = {x//y}")
    return inner(x, y)

print(f"Nested loop is created : {outer(100, 25)}")
```

```python
# Returning multiple values 
def multi_input(x):
    return f"add = {x+x}", f"minus = {x-x}", f"square = {x**2}", f"cube = {x**3}", f"devide = {x//x}"

print(multi_input(5)); print(multi_input(11)); print(multi_input(-1))
```

```python
# factorial !
def factorial(i):
    if i == 0:
        return 1
    else :
        return i * factorial(i-1)
print(factorial(5))
```

```python
# lambda function ex:
minus = lambda x, y:x-y
print(minus(3, 5))


oprations = lambda a, b, c, d, e, f: (a+b-c)*d//e//f
print(oprations(1, 2, 3, 4, 5, 6))
```

```python
# map function ex:
num = {2.25, 3, 6, 7, 8, 9, 0.0, 10, 24}
print(f"map func : {sorted(set(map(lambda x:x**3, num)))}")

# map with filter function
z= set(filter(lambda x:x%2!=0, num))
print(f"filter func : {sorted(z)}")
print(f"map + filter : {sorted(set(map(lambda x:x**4, z)))}")
```
