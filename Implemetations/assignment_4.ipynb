```python
# converting and appends 
tpl = tuple(range(1,5))
print(f"tpl_original {tpl}")
lst = list(tpl)
lst.append(15)
print(f"tpl_2_lst {lst}")
tpl = tuple(lst)
print(f"new_tpl {tpl}")
```

```python
# tuple with dict
dict_tpl = {
    ("hii", "hello"): "Oyee",
    ("fine ?", "okky ?"): ".",
    (str(1)+"st", str(2)+"nd"): "Bro"
}
print(dict_tpl)
for i in dict_tpl:
    print(f"{i} : {dict_tpl.values()}")
```

```python
# access each elment of a tuple matrix

mat_tpl = (
    ("hii", "hello", 0),
    ("fine ?", "okky ?", 1),
    (str(1)+"st", str(2)+"nd", 0.0)
)

for i in mat_tpl:
    for j in i:
        print(j)
```

```python
# make an random tuple and find their min, max, sum of element
def min_tpl(i):
    return min(i)

def max_tpl(i):
    return max(i)

def sum_tpl(i):
    return sum(i)


import random

random_tpl = tuple(random.randint(1, 100) for _ in range(20))

print(random_tpl)
print(f"Min value is {min_tpl(random_tpl)}")
print(f"Max value is {max_tpl(random_tpl)}")
print(f"Sum is {sum_tpl(random_tpl)}")
```
