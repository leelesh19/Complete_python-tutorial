```python
# Why function ?
# We can make our own function for specific code 
# Ex - find odd/even
def even_odd(n):
    """This is my own function for odd/even"""
    if n%2==0:
        print(f"{n} is even num.")
    else :
        print(f"{n} is odd num.")
```

```python
# How function work
even_odd(-38)
even_odd(8)
even_odd(87)
even_odd(0.0)
even_odd(-2)
even_odd(-10) 

# We can reuse the function code not to write same code again
```

```python
# function with multiple parameters
def percent(a, b, c, d, e):
    sum_percent = ((a+b+c+d+e)*100)/500
    return sum_percent

percent(88, 91, 87, 95, 94)
```

```python
# default function
def legend(name="User"):
    print(f"Hey {name}, welcome to the universe !")

legend()
legend("Guys")
```

```python
# Positional argument (* for multi parametes)
def positional_args(*para):
    print("This is how positional args work :")
    for i in para:
        print(f" - {i}")


positional_args(1, 53, 69, 12, 0.0, 'hey', 'hii', 'oyee...', '.', 'hira', 'are you fine ?', 'is everything okk ?', 'koi na, life hai chalta hai', 'I am with you.')
```

```python
# keyword args (** for dict form)

def keyword_args(**dict):
    print("Inrodusing some unknown one :")
    for key, value in dict.items():
        print(f"- {key} : {value}")

keyword_args(name='unknown', wanna='explore I', eligible='wanna worthy', skill='overthinking soul', fear='being hurt', feels='alone', strenght='wanna be', help_of='curiousness') 
```

```python
def multi_args(*args, **dict):
    print("This is how positional args work :")
    for i in args:
        print(f"Positional Args - {i}")


    print("Inrodusing some unknown one :")
    for key, value in dict.items():
        print(f"- {key} : {value}")
```
