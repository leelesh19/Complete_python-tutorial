```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import pandas as pd 
import seaborn as sns
```

```python
cars = sns.load_dataset("mpg")
cars.head()
```

```python
cars.describe()
```

```python
np.percentile(cars["mpg"], [25, 50, 75, 99, 100])
```

```python
#boxploting
sns.boxplot(cars["mpg"])
plt.title("Boxplotof mpg")
plt.show()  # ????? - solved
```

```python
# CHARTS
sns.countplot(cars["origin"])  
plt.title("Charts of origin")
plt.show()  # ????? - solved## gies charts
```

```python
sns.displot(cars["horsepower"]) 
plt.show()
```

```python
# Customized countplot for 'mpg' -> gives chart of it 
sns.countplot(x="mpg", data=cars, palette="Set2", edgecolor="black")
plt.xlabel("X-axis", fontsize=8)
plt.ylabel("Y-axis", fontsize=13)
plt.title("Charts", fontsize=20)
plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(x="mpg", y="horsepower", data=cars)
plt.xlabel("miles per gain (mpg)")
plt.ylabel("horsepower")
plt.title("the comparition of two charts")
plt.show()
```
