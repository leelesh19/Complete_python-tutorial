## Basic Calculator

```python
## let's make simple calculator program 
num1= float(input("Enter the first number: "))
num2= float(input("Enter the second number "))

sum = num1 + num2
difference = num1 - num2 
product = num1 * num2
devide = num1 / num2
remender = num1 % num2
power = num1 ** num2 

print("sum:", sum)
print("difference:", difference)
print("product:", product)
print("devide:", devide)
print("remender:", remender)
print("power:", power)     #here we gives input as first- 5 , second- 2
```
