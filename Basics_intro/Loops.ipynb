## Loop in python

```python
for i in range(10, 20):
    print(i)
```

```python
for i in range(3, -50, -5):
    print(i)
```

```python
# with string
str="lakhan lal yadav"
for i in str:
    print(i)
```

```python
# while loop 
count = -5

while count%2==0:
    print(count)

while (count%5==0) and (count>-10000):
    print(count)
    count = count*2
```

```python
# Loop controll statement break and continue
for i in range(2,20):
    if i>=15:
        break
    if i%2==0:
        continue
    print(i)
```

```python
# Nested loop 
for i in range(5):
    for j in range(2):
        print(f"i : {i} and j : {j} = {i and j}")
```

## Calculate the sum of first natural num using loop

```python
n = int(input("0-n first num sum : ")) # we can directly put value of n

sum=0
count=1

while count<=n:
    sum = sum + count 
    count = count + 1 
    print(f"The sum of first {count-1} natural num is : {sum}")
```

```python
n=int(input("Enter a natural num : "))
sum=0
for i in range(n+1):
    sum = sum + i 
print(f"The sum of first {n} natural num is : {sum}")
```

```python
# find all prime num between 1 - 100 

for num in range(1, 101):
    if num>1:
        for i in range(2, num):
            if num%i==0:
                break
        else : 
            print(num)
```
