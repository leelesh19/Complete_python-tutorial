```python
# find odd, even, +ve , or -ve num
num = int(input("Enter a number : "))

print(f"num is {num}")
if num>=0:
    print("It's a positive number")
    if num % 2==0:
        print("It's an even number")
    else :
        print("It's an odd number")
else : 
    print("It's zero or negative number")
```

```python
# find a leap year 
year = int(input("Enter a year : "))

if year % 4 ==0:
    if year % 100==0:
        if year % 400==0:
            print(year, "is a leap yaer")
        else :
            print(year, "is not a leap year")
    print(year, "is a leap year")
else :
    print(year, "is not a leap year")

    
```

```python
## Simple calculator 
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
opration = input("Enter the opretion (+, -, *, **, /, %) : ")


if opration == "+":
    result = num1 + num2 
elif opration=="-":
    result = num1 - num2 
elif opration=="*":
    result = num1 * num2 
elif opration=="/":
    result = num1 / num2
elif opration=="**":
    result = num1 ** num2 
elif opration=="%":
    result = num1 % num2
print("first : ", num1)
print("second :", num2)
print("opretion :", opration)
print(result)
```

```python
## Determine the ticket price by age 

age = int(input("Enter your age : "))
is_student = input("Are you a student ? (yes/no):")

if age<5:
    price = "free"
elif age<=10:
    price = "$10"
elif age<=15:
    if is_student=="yes":
        price = "$12"
    else :
        price = "$15"
elif age<=18:
    if is_student=="yes":
        price = "$20"
    else :
        price = "$30"
elif age<=40:
    if is_student=="yes":
        price = "$40"
    else :
        price = "$50"
elif age>40: 
    if is_student=="yes":
        price = "$100"
    else :
        price = "$1K"

print("ticket price :", price)
```
