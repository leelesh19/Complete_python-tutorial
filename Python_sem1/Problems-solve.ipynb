#### 1. WAP to find roots of a quadric equation.

```python
def roots(a, b, c):
    d = b*b - 4*a*c
    if d>0:
        root1= (-b+(d**0.5))/2*a
        root2= (-b-(d**0.5))/2*a
        print("roots are real and different:", "root1 =",root1, "root2 =", root2)
    elif d==0:
        root = -b/(2*a)
        print("roots are real nd equal :", root)
    elif d<0:
        img = (d**0.5)/2*a
        print("roots are complex and imaginary :", "img =", img)

a = float(input("enter value of a :"))
b = float(input("enter value of b :"))
c = float(input("enter value of c :"))

roots(a, b, c)
```

#### 2. WAP to accept a number "n" :

(i) Check if 'n' is prime-

```python
n=int(input("enter a num :"))

if n>1:
    for i in range(2, n):
        if n%i==0:
            print(n,'is not a prime.')
            break
        else :
            print(n, 'is a prime.')
            break
else :
    print("Number should be greater than 1.")
```

(ii) Generate all prime till 'n'-

```python
n = int(input("enter a num :"))

print(f"all prime till {n} are :")

for num in range(2, n+1):
    for i in range(2, int(num**0.5)+1):
        if num % i ==0:
            break
    else:
        print(num, end=" ")  # here else : is not in if loop it's connected to for loop
```

(iii) Generate first "n" prime number-

```python
n = int(input("enter a num :"))
num = 2
count =0
print(f"first {n} prime numbers are :")
while n>count:
    for i in range(2, int(num**0.5)+1):
        if num%i==0:
            break
    else :
        print(num, end=",")
        count +=1
    num+=1
```

#### 3. WAP to create a pyramid of the character "*" and a reverse pyramid

(i) pyramid made of : "*"

```python
n = int(input("enter hieght of pyramid :"))
m=1
for i in range(65, 71):
    space = " "*(n)
    stars = chr(i)*m
    print(space+stars) 
    n-=1
    m+=2
```

(ii) reverse pyramid made of : "*"

```python
n = int(input("enter hieght of inverted pyramid :"))

for i in range(n, 0, -1):
    space = " "*(n-i)*2
    stars = "* "*(2*i - 1)
    print(space+stars) 
```

(iii) diamond shape made of : "*"

```python
n = int(input("enter hieght of diamond shape :"))
#upper pyramid
for i in range(1, n+1):
    for j in range(65, 70):
        space = ' '*(n-i)*2
        stars = chr(j)*(2*i-1)
        print(space+stars)
#lower inverted pyramid
for j in range(n-1, 0, -1):
    space=' '*(n-j)*2
    stars= "* "*(2*j-1)
    print(space+stars)
```

#### 8. WAP to create a list of cubes of only the even integers appearing in the input list (may have elements of other types also) using the following :

(a) using "for" loop

```python
items = [2, 5, 'hello', 8, 3.5, 10, 'python', 7, 0, True, 12, 20.00]     # it is an list which content other types of elements also

even_cubes = []

for i in items:
    if type(i) == int and i%2 == 0:
        even_cubes.append(i**3)

print("list of even cube is :", even_cubes)

```

(b) using list comprehension

```python
items = [2, 5, 'hello', 8, 3.5, 10, 'python', 7, 0, True, 12, 20.00]     # it is an list which content other types of elements also

even_cubes = [i**3 for i in items if type(i)==int and i%2==0]

print("list of even cube is :", even_cubes)
```

#### 10. WAP to define a class Point with coordinates x and y as attributes. Create relevant methods and print the objects. Also define a method distance to calculate the distance between any two point objects.

```python
import math

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def display(self):
        print(f"Point({self.x}, {self.y})")

    def distance(self, other):
        # Distance formula: √[(x2 - x1)² + (y2 - y1)²]
        return math.sqrt((other.x - self.x)**2 + (other.y - self.y)**2)

# Creating point objects
p1 = Point(3, 4)
p2 = Point(7, 1)

# Printing the points
p1.display()
p2.display()

# Calculating distance
d = p1.distance(p2)
print("Distance between points:", d)
```
